```python
import matplotlib.pyplot as plt

def determinar_relacion_rectas(x1, y1, x2, y2, x3, y3, x4, y4):
    m1 = (y2 - y1) / (x2 - x1)  # Pendiente de la primera recta
    m2 = (y4 - y3) / (x4 - x3)  # Pendiente de la segunda recta

    if m1 == m2:
        return "Las rectas son paralelas."
    elif m1 * m2 == -1:
        return "Las rectas son perpendiculares."
    else:
        return "Las rectas no son paralelas ni perpendiculares."

def obtener_ecuacion_recta(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)  # Pendiente
    b = y1 - m * x1  # Término independiente
    return f"y = {m}x + {b}"

# Obtener las coordenadas de los puntos para cada recta
x1 = float(input("Ingrese la coordenada x del primer punto de la primera recta: "))
y1 = float(input("Ingrese la coordenada y del primer punto de la primera recta: "))
x2 = float(input("Ingrese la coordenada x del segundo punto de la primera recta: "))
y2 = float(input("Ingrese la coordenada y del segundo punto de la primera recta: "))

x3 = float(input("Ingrese la coordenada x del primer punto de la segunda recta: "))
y3 = float(input("Ingrese la coordenada y del primer punto de la segunda recta: "))
x4 = float(input("Ingrese la coordenada x del segundo punto de la segunda recta: "))
y4 = float(input("Ingrese la coordenada y del segundo punto de la segunda recta: "))

# Determinar la relación entre las rectas
relacion = determinar_relacion_rectas(x1, y1, x2, y2, x3, y3, x4, y4)
print(relacion)

# Obtener la ecuación de cada recta
ecuacion_recta1 = obtener_ecuacion_recta(x1, y1, x2, y2)
ecuacion_recta2 = obtener_ecuacion_recta(x3, y3, x4, y4)
print("Ecuación de la primera recta:", ecuacion_recta1)
print("Ecuación de la segunda recta:", ecuacion_recta2)

# Graficar las rectas
x = [x1, x2, x3, x4]
y = [y1, y2, y3, y4]
plt.plot(x[:2], y[:2], label='Recta 1')
plt.plot(x[2:], y[2:], label='Recta 2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de las rectas')
plt.legend()
plt.show()
```
